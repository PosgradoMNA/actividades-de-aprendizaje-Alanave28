#####**Box Plots**

Son la mejor manera de visualizar datos numéricos y permiten visualizar distintas distribuciones de datos.

- Estos diagramas permiten ver la distribución y asimetría de los datos.

- Facilitan la comparación entre grupos de datos.

####**Análisis de Varianza**
- Analizar una variable categórica y ver la correlación entre diferentes categorías.


ANOVA - ANalysis Of VAriance
- Encuentra la correlacion entre diferentes grupos de una variable categórica.
- Devuelve 2 valores:
  * F-test score - calcula la relación de varianza entre la media de los grupos de variación.
  (Relación de variación entre medias de grupo sobre la variación dentro de cada una de las medias del grupo de muestra).
  * P-value - muestra si el resultado obtenido es estadísticamente significativo.

Ejemplo con Scipy:

    * dfAnova = df[['col1', 'col2$']]
    * dfGpAnova = dfAnova.groupby(['col1'])
    * resAnova = stats.f_oneway(dfGpAnova.get_group('val1')['col2$'], dfGpAnova.get_group('val2')['col2$'])

Existe una fuerte correlación entre una variable categórica y otras variables cuando ANOVA nos da un valor F grande y una valor P pequeño.

####**Correlación**

Es una mpetrica estadística para medir en que medida las diferentes variables son interdependientes.

Ejemplo: 
- Cuando observamos dos variables a lo largo del tiempo, si una variable cambia, ¿cómo este efecto cambio en la otra variable?.

La correlación no implica causalidad.

Se llega a utilizar una gráfica de dispersión con una línea adicional para visualizar la correlación entre 2 variables.

    * sns.regplot(x="col1", y="col2$", data=df)
    * plt.ylim(0,)

  

####**Métodos estadísticos de Correlación**

Una forma de medir la fuerza de correlación entre variables númericas continuas es la Correlación Pearson, esta devuelve 2 valores:

    * Coeficiente de correlación
      - Valor cercano a 1 = Gran correlación Positiva
      - Valor cercano a -1 = Gran correlación Negativa
      - Valor cercano a 0 = No hay correlación entre variables

    * P-value
      - Valor inferior a 0,001 = Gran certeza sobre la correlación
      - Valor entre 0,001 y 0,05 = Certeza moderada
      - Valor entre 0,05 y 0,1 = Certeza débil 
      - Valor mayor que 0.1 = Ninguna certeza de la correlación

Existe una fuerte correlación cuando el coeficiente de correlación es cercano a 1 o -1 y el valor p es menor que 0.001.

Ejemplo

    * pearson_coef, p_value = stats.pearsonr(df['col1'], df['col2'])
