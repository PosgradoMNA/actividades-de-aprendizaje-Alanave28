###**Desarrollo de modelos**

####**Modelo**

Se puede pensar un modelo como una ecuación matemática que se utiliza para:
- Predicir uno o más valores dados.
- Relacionar una o más variables o características independientes con variables dependientes.

Cuantos más datos relevantes tenga un modelo, este será más preciso.



###**Regresión lineal simple y múltiple**

Regresión lineal
- Variable independiente para realizar una predicción.

Regresión lineal múltiple
- Múltiples variables independientes para realizar una predicción.

####**Regresión linel simple (SLR)**
- Es un método que ayuda a entender la relación entre 2 variables:
  * Variable predictora (X, independiente)
  * Variable objetivo (y, dependiente)

Para determinar una relación linea
- Determinar línea
  * Tomar puntos de datos de nuestro conjunto de datos.
  * Se utilizan estos punto de entrenamiento para ajustar el modelo.
  * El resultado se le conoce como parámetro.
  * Estos se almacenan en dos macros de datos o en matrices numpy.
  * El valor a predecir -> Se almacena en la matriz y.
  * La variable dependiente -> Se almacena en la matriz x.
  * Cada muestra corresponde a una fila de la matriz.

Se resume el proceso:
- Se tiene un conjunto de datos de entrenamiento.
- Estos se utilizan para ajustar o entrenar el modelo y obtener parámetros.
- Estos parámetros se utilizan en el modelo.
- Obteniendo este modelo, la función se convierte en una estimación.
- El modelo se utiliza para predecir valores que no se han visto.

El modelo no siempre es correcto, porque se valida vs el valor real.
- Si la suposición lineal es correcta, el error se debe al ruido u otras razones.

Para ajustar el modelo:
- Se importa el modelo linea de Scikit-learn
  * from sklearn.linear_model import LinearRegression

- Se crea un objeto de la Regresión lineal utilizando el constructor
  * lm = LinearRegression()

- Se define la variable predictora y la variable objetivo
  * X = df['col1']
  * Y = df['col2']

- Se utiliza el método fit para ajustar el modelo y encontrar los valores b0 y b1
  * lm.fit(X, Y)
  * Yhat =  lm.predict(X)

- La salida es una matriz con el mismo número de muestras que la entrada X.

- El intercept (b0) y el slope (b1) son atributos del objeto lm.
  * lm.intercept_
  * lm.coef_


####**Regresión linel múltiple (MLR)**
Se utiliza para explicar la relación entre una variable objetivo continua (Y) y dos o más variable predictoras (X).

Para ajustar el modelo:
- Se extraen las variables predictoras y se almacenan en una variable.
  * Z = df[['col1', 'col2', 'col3', 'col4']]

- Se entrena como en el ejemplo anterior

###**Regresión polinomial y pipelines**

- Se transforman los datos en un polinomio, luego aplicamos la regresión lineal para ajustar el parámetro.

- Caso especial de la regresión lineal general.

- Este método es perfecto para describir relaciones curvilíneas.
  * La relación curvilínea es la que se obtiene elevando al cuadrado o estableciendo términos de orden superior de las variables predictoras
en el modelo, transformando los datos.

  * El modelo puede ser cuadrático, lo que significa que la variable predictora del modelo está elevada al cuadrado.

- Regresión polinomial de segundo orden
  * Y = b0 + b1x1 + b2(x1)^2

- Regresión polinomial de tercer orden
  * Y = b0 + b1x1 + b2(x1)^2 + b3(x1)^3

- Regresión polinomial de ordern superior
  * Y = b0 + b1x1 + b2(x1)^2 + b3(x1)^3 ...

- El grado de regresión hace una gran diferencia y puede resultar en un mejor ajuste si elige el valor correcto.

- La relación entre la variable y el parámetro es siempre lineal.

Ejemplo:

    * f = np.polyfit(x,y,3)
    * p = np.polyld(f)
    * print(p) -> Se imprime el modelo

####**Pipelines**

- Pipelines son una forma de simplificar el código.

- Realizan secuencialmente una serie de transformaciones.
  * Normalización
  * Transformación polinomial

- El último paso realiza la predicción.
  * Regresión Lineal

- Constructor
  * Input = [('scale', StandardScaler()), ('polynomial', PolynomialFeatures(degree=2)), ('model', LinearRegression())]

- Se puede entrenar utilizando método "train"

Ejemplo: 

    * from sklearn.preprocessing import PolynomialFeatures
    * from sklearn.linear_model import LinearRegression
    * from sklearn.preprocessing import StandardScaler
    * from sklearn.pipeline import Pipeline

    * Input = [('scale', StandardScaler()), ('polynomial', PolynomialFeatures(degree=2)), ('model', LinearRegression())]

    * pipe = Pipeline(Input)

    * Pipe.train(X['col1','col2', 'col3', 'col4'], y)

Pasos para obtener una predicción: 

    * Normalización
    * Transformación polinomial
    * Regresión lineal

###**R-square y MSE para evaluación dentro de la muestra**

- Estas medidas son una forma de determinar numéricamente qué tan bien se ajusta el modelo a nuestros datos.

- Las 2 medidas más utilizadas para determinar el ajuste del modelo:
  * Error cuadrado medio (MSE)
  * R-cuadrado (R^2)

####**MSE**
Se determina:
- Diferencia entre valor real y valor predicho
- Resultado se eleva al cuadrado
- Se toma la media o el promedio de todos los errores (se suman todos y se divide por el número de muestras).

- En python mean_squared_error obtiene 2 entradas
  * Valor real 
  * valor predicho

Python

    * from sklearn.metrics import mean_squared_error
    * mean_squared_error(df['col1'], Y_Predict_fit)


####**R-squared**

- También llamado coeficiente de determinación.
- Es una medida para determinar qué tan cerca están los datos de la línea de regresión ajustada.


Se determina:

    * R^2 = (1- (MSE Regresion Lineal / MSE del promedio de los datos))

Python

    * X = df['col1']
    * Y = df['col2']
    * lm.fit(X, Y)
    * lm.score(X, y)

- El valor normal de R^2 debe estar entre 0 y 1.
- Si R^2 es negativo, puede deberse a un sobreajuste.

####**3. Polynomial Regression and Pipelines**

```python
def PlotPolly(model, independent_variable, dependent_variabble, Name):
    x_new = np.linspace(15, 55, 100)
    y_new = model(x_new)

    plt.plot(independent_variable, dependent_variabble, '.', x_new, y_new, '-')
    plt.title('Polynomial Fit with Matplotlib for Price ~ Length')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Price of Cars')

    plt.show()
    plt.close()
```

```python
x = df['highway-mpg']
y = df['price']
f = np.polyfit(x, y, 3)
p = np.poly1d(f)
print(p)
```

```python
PlotPolly(p, x, y, 'highway-mpg')
```

```python
np.polyfit(x, y, 3)
```

```python
#Pregunta 4 - Create 11 order polynomial model with the variables x and y from above.
f1 = np.polyfit(x, y, 11)
p1 = np.poly1d(f1)
print(p1)
PlotPolly(p1,x,y, 'Highway MPG')
```
